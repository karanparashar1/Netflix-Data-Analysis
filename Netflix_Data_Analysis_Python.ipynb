## Data Information

```python
def initial_EDA():     #function for printing basic information about the data.
    print("Summary \n\n", data.info())     # Column name, number of non-empty in it amd their data type.
    print("\n--------------------------------------------\n")
    print("Shape\n\n", data.shape)     # Shape of the data frame (Number of content and columns in it).
    print("\n--------------------------------------------\n")
    print("Data Types\n\n", data.dtypes)     # Data types of each column.
    print("\n--------------------------------------------\n")
    print("Any Duplicate Values?\n\n", data.duplicated().sum())     # NUmber of duplicate entries.
    print("\n--------------------------------------------\n")
    print("Empty Values Total \n\n", data.isnull().sum())     # Number of empty value in each column.
    print("\n--------------------------------------------\n")
    print("Non-empty values Total \n\n", data.notnull().sum())     # Number of non-empty value in each column.
    print("\n--------------------------------------------\n")
    print("Columns names\n\n", data.columns)     # Name of each column.
    print("\n--------------------------------------------\n")
    print("Unique Values\n\n", data.nunique())     # Number of uniques value in each column.
    print("\n--------------------------------------------\n")
initial_EDA()
```

# Netflix Search On Google

```python
kw_list=["Netflix"]     # list of keywords to get data.
trends.build_payload(kw_list)     # The build_payload method from Pytrends is used to build a list of keywords that we want to search in Google Trend.
data = trends.interest_by_region()     # Using interest_by_region() which gives the data of the list of keyword over countries.
data = data.sort_values(by="Netflix", ascending=False)     # Sorting the data in decreasing values of relative search.  
data = data.head(20)     # Storing only the first 20 countries. 
print(data)
```

```python
data.reset_index().plot(x="geoName", y="Netflix", figsize=(20,15), kind="bar", title = 'Interest by region')
# plt.style.use('classic')
plt.show()
```

```python
data = TrendReq(hl="en-US", tz=360)     # Calling the Pytrends API. 
data.build_payload(kw_list)     # The build_payload method from Pytrends is used to build a list of keywords that we want to search in Google Trend.
data = data.interest_over_time()     # Using interest_over_time() method which gives the data of the list of keyword over over time.
fig = plt.subplots(figsize = (20,15))     # For resizing the plot.
data['Netflix'].plot()     # Plotting the graph for 'Netflix' column from the data. 
plt.title('Total Google Searches for Netflix', fontweight='bold')     
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

```python
data_update = data.groupby('Country').count()  

fig = px.choropleth( data_update,
                     locations = data_update.index,
                     locationmode = 'country names',
                     scope = 'world',
                     title = 'Countries whose content is available on Netflix')
fig.show()
```
